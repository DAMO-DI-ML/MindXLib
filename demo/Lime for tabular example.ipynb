```python
import numpy as np

import sklearn
from sklearn import datasets, ensemble
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression

from mindxlib.post_hoc_explainer import LimeTabularExplainer
```

```python
iris = datasets.load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, train_size=0.8)

np.random.seed(42)
rf = ensemble.RandomForestClassifier(n_estimators=200)
rf.fit(X_train, y_train)
i = np.random.randint(0, X_test.shape[0])
```

```python
feature_names = iris.feature_names
target_names = iris.target_names

explainer = LimeTabularExplainer(X_train, mode="classification", feature_names=feature_names,
                                class_names=target_names, discretize_continuous=True)
exp = explainer.explain_instance(X_test[i], rf.predict_proba, num_features=2, model_regressor=LinearRegression())
```
